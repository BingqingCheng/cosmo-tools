```python
def cbarPoly():
    x = np.linspace(0.0, 1.0, 100)
    r = np.array([28.384, -66.085, 54.494, -16.557, -0.232, 0.991])
    g = np.array([4.196, -8.275, 9.143, -5.479, -0.086, 0.995])
    b = np.array([-2.312, 4.468, -3.085, -0.128, 0.052, 0.997])
    pr = np.polyval(r, x)
    pg = np.polyval(g, x)
    pb = np.polyval(b, x)
    for i in [pr, pg, pb]:
        i[np.where(i > 1.0)] = 1.0
        i[np.where(i < 0.0)] = 0.0
    cbpoly = LinearSegmentedColormap.from_list('cbarPoly', np.column_stack((pr, pg, pb)))
    plt.register_cmap(cmap=cbpoly)
    
    colors = cbpoly(np.arange(cbpoly.N))
    figColormap = plt.figure(figsize=(6.7, 3.4)) 
    cAx = figColormap.add_subplot(1, 1, 1)
    cAx.imshow([colors], extent=[0, 10, 0, 1])
    cAx.set_xticklabels([])
    cAx.set_xticks([])
    cAx.set_yticklabels([])
    cAx.set_yticks([])
    plt.show()

def cbarHot():
    x = np.linspace(0.0, 1.0, 1000)
    y = np.linspace(0.0, 0.8, 1000)
    rgbWhite = np.array([1.0, 1.0, 1.0])*np.ones((1000, 3))
    rgb1 = np.array([1.0, 0.5, 0.0])*np.ones((1000, 3))
    rgbPurple = np.array([0.5, 0.0, 0.5])*np.ones((1000, 3))
    cm = plt.get_cmap('ocean')
    dsc = cm(1.0-y)**1.1
    c1 = (1.0-x)**4
    c2 = x**10
    c3 = (x-0.1)**2
    blend1 = np.einsum('ij,i->ij', dsc[:, 0:3], 1.0-c1) + np.einsum('ij,i->ij', rgbWhite, c1)
    blend3 = np.einsum('ij,i->ij', blend1, 1.0-c3) + np.einsum('ij,i->ij', rgbPurple, c3)
    blend2 = np.einsum('ij,i->ij', blend3, 1.0-c2) + np.einsum('ij,i->ij', rgb1, c2)
    
    cbh = LinearSegmentedColormap.from_list('cbarH', blend2)
    plt.register_cmap(cmap=cbh)
    
    colors = cbh(np.arange(cbh.N))
    figColormap = plt.figure(figsize=(6.7, 3.4)) 
    cAx = figColormap.add_subplot(1, 1, 1)
    cAx.imshow([colors], extent=[0, 10, 0, 1])
    cAx.set_xticklabels([])
    cAx.set_xticks([])
    cAx.set_yticklabels([])
    cAx.set_yticks([])
    plt.show()
    
cbarPoly()
cbarHot()
cbh = cosmocbars.cbarHot()
cbh.view()
x = plt.get_cmap('cbarHot')
y = plt.get_cmap('cbarPoly')

print x(0.2)
print y(0.2)
```

```python
def cbarBWR():
    # RdBu matplotlib color bar recommended instead
    x = np.linspace(0.0, 1.0, 1000)
    rgbBlack = np.array([0.0, 0.0, 0.0])*np.ones((1000, 3))
    rgbWhite = np.array([1.0, 1.0, 1.0])*np.ones((1000, 3))
    rgb1 = np.array([0.0, 0.4, 1.0])*np.ones((1000, 3))
    rgb2 = np.array([1.0, 0.3, 0.0])*np.ones((1000, 3))
    c1 = np.sin(x*np.pi/2)
    c2 = np.cos(x*np.pi/2)
    c3 = 2*np.sin(x*np.pi)**4 + 4*np.sin(x*np.pi)**16 + 8*np.sin(x*np.pi)**64 + 32*np.sin(x*np.pi)**256
    c3 /= np.amax(c3)
    c4 = np.cos(x*np.pi)**16
    norm = np.amax(c1+c2+c3+c4)
    c1 /= norm
    c2 /= norm
    c3 /= norm
    c4 /= norm
    blend = np.einsum('ij,i->ij', rgb2, c1) + np.einsum('ij,i->ij', rgb1, c2) \
        + np.einsum('ij,i->ij', rgbWhite, c3) + np.einsum('ij,i->ij', rgbBlack, c4)
    
    cbwr = LinearSegmentedColormap.from_list('cbarBWR', blend)
    plt.register_cmap(cmap=cbwr)
    
    colors = cbwr(np.arange(cbwr.N))
    figColormap = plt.figure(figsize=(6.7, 3.4)) 
    cAx = figColormap.add_subplot(1, 1, 1)
    cAx.imshow([colors], extent=[0, 10, 0, 1])
    cAx.set_xticklabels([])
    cAx.set_xticks([])
    cAx.set_yticklabels([])
    cAx.set_yticks([])
    plt.show()
    
cbarBWR()
```

```python
def cbarPhi():
    x = np.linspace(0.0, 1.0, 1000)
    rgbBlack = np.array([0.0, 0.0, 0.0])*np.ones((1000, 3))
    rgbWhite = np.array([1.0, 1.0, 1.0])*np.ones((1000, 3))
    rgb1 = np.array([0.0, 0.4, 0.9])*np.ones((1000, 3))
    rgb2 = np.array([1.0, 0.3, 0.0])*np.ones((1000, 3))
    c1 = 0.5*np.cos(x*np.pi+np.pi/4+np.pi/4)**4
    c2 = 1.0*np.sin(x*np.pi+np.pi/4+np.pi/4)**8
    c3 = np.cos(x*np.pi+np.pi/4)**4
    c4 = np.cos(x*np.pi+np.pi/2+np.pi/4)**4
    norm = np.amax(c1+c2+c3+c4)
    c1 /= norm
    c2 /= norm
    c3 /= norm
    c4 /= norm
    blend = np.einsum('ij,i->ij', rgbBlack, c1) + np.einsum('ij,i->ij', rgbWhite, c2) \
        + np.einsum('ij,i->ij', rgb1, c3) + np.einsum('ij,i->ij', rgb2, c4)

    cbp = LinearSegmentedColormap.from_list('cbarPhi', blend)
    plt.register_cmap(cmap=cbp)
    
    rotation = np.arange(0, 361, 1)
    thickness = np.arange(40, 70, 1)
    values = rotation*np.ones((30, 361))
    fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
    ax.pcolormesh(rotation*np.pi/180.0, thickness, values, cmap=plt.get_cmap('cbarPhi'))
    ax.set_yticks([])
    ax.tick_params(pad=15)
    plt.show()
    
cbarPhi()
```
